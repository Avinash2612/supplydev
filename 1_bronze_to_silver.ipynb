```python
# 1. Point to the FOLDER where ADF drops the API files, instead of a specific file name
# (Replace with your actual account name and bronze container folder path)
bronze_folder_path = "abfss://bronze@saadbprojectsupply.dfs.core.windows.net/raw_api_adf/" 

# By pointing to the folder, Spark will read all JSON files inside it together
bronze_df = spark.read.format("json").load(bronze_folder_path)

# 2. Cleanse and enrich the data
from pyspark.sql.functions import current_timestamp
silver_df = bronze_df.withColumn("ingestion_timestamp", current_timestamp())

# 3. Write data to your registered Silver External Location
silver_delta_path = "abfss://silver@saadbprojectsupply.dfs.core.windows.net/cleaned_api_adf/"
silver_df.write.format("delta").mode("append").save(silver_delta_path)

print("Bronze folder successfully processed and appended to Silver!")
```
