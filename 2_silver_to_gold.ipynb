```python
# 1. Read the cleansed data from your registered Silver External Location Folder
# (Update with your actual storage account name)
silver_delta_path = "abfss://silver@saadbprojectsupply.dfs.core.windows.net/cleaned_api_adf"
silver_df = spark.read.format("delta").load(silver_delta_path)

# 2. Perform a business aggregation (e.g., counting total posts per user)
# This creates a small, highly fast table optimized for your Power BI visuals
from pyspark.sql.functions import count
gold_df = silver_df.groupBy("userId").agg(count("id").alias("total_posts"))

# 3. Write data to your Gold External Location
# We use 'overwrite' mode so that every time this runs, Power BI sees the latest data
gold_delta_path = "abfss://gold@saadbprojectsupply.dfs.core.windows.net/posts_by_user_summary"
gold_df.write.format("delta").mode("overwrite").save(gold_delta_path)

print("Silver to Gold complete! Table is optimized and ready for Power BI connection.")
```
